```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import rc
import matplotlib as mpl
import matplotlib.ticker as ticker
from scipy.stats import norm
```

```python
# Matplotlib customize
plt.rcParams['figure.figsize'] = [6,5]
plt.rcParams['figure.dpi'] = 100

plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 20

plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['lines.markersize'] = 5

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

plt.rcParams['legend.fontsize'] = 15
plt.rcParams['legend.markerscale'] = 1.5
plt.rcParams['legend.borderpad'] = 0.6
plt.rcParams['legend.framealpha'] = 0.7
```

```python
x = np.linspace(-10,10,100)
y0 = norm.pdf(x,loc = 3, scale=2)
y1 = norm.pdf(x,loc = -1, scale=2.1)
QObs = -0.5
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y0,color='green',label=r'$f(\mathcal{Q},0)$')
ax.plot(x,y1,color='yellow',label=r'$f(\mathcal{Q},1)$')
ax.axvline(x = QObs, color='k', ls='--',label=r'$\mathcal{Q}_{observed}$')

# Pvalues
p0scan = np.linspace(x[0],QObs,100)
ax.fill_between(p0scan,0,norm.pdf(p0scan,loc = 3, scale=2.),color='green',label=r'$p_{0}$')

p1scan = np.linspace(QObs,x[-1],100)
ax.fill_between(p1scan,0,norm.pdf(p1scan,loc = -1, scale=2.1),color='yellow',label=r'$p_{1}$')

ax.set_xlabel(r'$\mathcal{Q}$')
ax.xaxis.set_label_coords(0.95,-0.1)
#ax.set_xlim(x[0],x[-1])

# Label Y
ax.set_ylabel(r'$p(\mathcal{Q};\mu)$')
ax.yaxis.set_label_coords(-0.15,0.85)
#ax.set_ylim(np.min(y),np.max(y))

#ax.set_yscale('log')

# Tickers
ax.tick_params(axis='both', which='major', labelsize=12)

plt.annotate('Signal like', xy=(-9, 0.05), xytext=(-6, 0.05),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Bkg like', xy=(9, 0.05), xytext=(3, 0.05),
             arrowprops=dict(facecolor='black', shrink=0.05))


# Ajustar los mrgenes automticamente
plt.tight_layout()

#Legend
ax.legend(loc='upper right',bbox_to_anchor=(1,1),ncol=1,title='')

# Saving figure
plt.savefig('LEP_estimator.pdf')
```
