```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import rc
import matplotlib as mpl
import matplotlib.ticker as ticker
from scipy.stats import norm
```

```python
# Matplotlib customize
plt.rcParams['figure.figsize'] = [6,5]
plt.rcParams['figure.dpi'] = 100

plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 20

plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['lines.markersize'] = 5

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

plt.rcParams['legend.fontsize'] = 15
plt.rcParams['legend.markerscale'] = 1.5
plt.rcParams['legend.borderpad'] = 0.6
plt.rcParams['legend.framealpha'] = 0.7
```

```python
mu = np.linspace(0.,3.,1000)
s = 10.
n = 105.
b = 100.
Q = lambda mu: 2*(mu*s - n*np.log(1+mu*s/b))
Q1 = Q(mu)
```

```python
minQ1 = np.min(Q1)
ii = np.where(Q1 == np.amin(Q1))
print(ii)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(mu,Q1,label=r'Log-likelihood ratio')

ax.set_xlabel(r'$\mu$')
ax.xaxis.set_label_coords(0.95,-0.1)
#ax.set_xlim(x[0],x[-1])

# Label Y
ax.set_ylabel(r'$-2Ln(\mathcal{Q}(\mu))$')
ax.yaxis.set_label_coords(-0.1,0.8)

ax.axhline(y = minQ1, color='r', linestyle='--')

# Ajustar los márgenes automáticamente
plt.tight_layout()

#Legend
ax.legend(loc='upper right',bbox_to_anchor=(1,1),ncol=1,title='')

# Saving figure
plt.savefig('LnQ.pdf')
```
