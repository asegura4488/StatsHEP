```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.ticker as ticker
from matplotlib import rc
from scipy.stats import chi2
```

```python
# Matplotlib customize
plt.rcParams['figure.figsize'] = [6,5]
plt.rcParams['figure.dpi'] = 100

plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 20

plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['lines.markersize'] = 4

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

plt.rcParams['legend.fontsize'] = 12
plt.rcParams['legend.markerscale'] = 1.5
plt.rcParams['legend.borderpad'] = 0.6
plt.rcParams['legend.framealpha'] = 0.7
```

```python
x = np.linspace(0,10,100)
n = b = 0
df = 2*(n+1)
y = chi2.pdf(x,df=df)
```

```python
# Upper limit calculation
upper = 0.5*chi2.ppf(0.95,df=df)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.plot(x,y,color='k', label=r'$\chi^{2}(2\mu; k=2(n+1))$')
# Con markers
#ax.plot(x,y,marker='o',color='k', label=r'$\chi^{2}(2\mu; k=2(n+1))$')

# Label X
ax.set_xlabel(r'$\mu$')
ax.xaxis.set_label_coords(0.95,-0.1)
ax.set_xlim(x[0],x[-1])

# Label Y
ax.set_ylabel(r'$p(\mu)$')
ax.yaxis.set_label_coords(-0.1,0.90)
ax.set_ylim(np.min(y),np.max(y))

ax.axvline( x = 2*upper, color='r', linestyle='--', label=r'$2\mu^{up}$= %.3f'%(2*upper) )
upper_ = np.linspace(2*upper,x[-1],100)
ax.fill_between( upper_, 0, chi2.pdf(upper_,df=df), color='b')

# Tickers
ax.tick_params(axis='both', which='major', labelsize=12)

#scale_ = 1
#ticks_x = ticker.FuncFormatter(lambda x,pos:'{:.1f}'.format(x/scale_)) 
#ax.xaxis.set_major_formatter(ticks_x)
#ticks_y = ticker.FuncFormatter(lambda y,pos:'{:.1f}'.format(y/scale_))
#ax.yaxis.set_major_formatter(ticks_y)

# Ajustar los márgenes automáticamente
plt.tight_layout()

#Legend
ax.legend(loc='upper right',bbox_to_anchor=(1,1),ncol=1,title='')

# Saving figure
plt.savefig('FrequentistUpperLimit.pdf')
```
