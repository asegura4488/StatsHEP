```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import rc
import matplotlib as mpl
import matplotlib.ticker as ticker
from scipy.stats import norm
```

```python
# Matplotlib customize
plt.rcParams['figure.figsize'] = [6,5]
plt.rcParams['figure.dpi'] = 100

plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 20

plt.rcParams['lines.linewidth'] = 1.5
plt.rcParams['lines.markersize'] = 5

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

plt.rcParams['legend.fontsize'] = 15
plt.rcParams['legend.markerscale'] = 1.5
plt.rcParams['legend.borderpad'] = 0.6
plt.rcParams['legend.framealpha'] = 0.7
```

```python
x = np.linspace(-6,6,100)
y = norm.pdf(x,loc = 0, scale=1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y,color='k',label=r'$\mathcal{N}(0,1)$')
ax.axvline(x = norm.ppf(0.95), color='g', ls='--',label=r'$P_{1-\alpha} = %.3f \sigma$' %(norm.ppf(0.95)))

I1 = np.linspace(norm.ppf(0.95),x[-1],50)
ax.fill_between(I1,0,norm.pdf(I1,loc = 0, scale= 1.),color='g',label=r'$\alpha = %.2f$' %(0.05))

p = 1 - norm.cdf(5,loc=0,scale=1)
print(p,1-p)

ax.axvline(x = norm.ppf(1-p), color='b', ls='--',label=r'$P_{1-\alpha} = %.1f \sigma$' %(norm.ppf(1-p)))

I2 = np.linspace(norm.ppf(1-p),x[-1],50)
ax.fill_between(I2,0,norm.pdf(I2,loc = 0, scale= 1.),color='b',label=r'$\alpha = %.2e$' %(p))


ax.set_xlabel(r'$x$')
ax.xaxis.set_label_coords(0.95,-0.1)
#ax.set_xlim(x[0],x[-1])

# Label Y
ax.set_ylabel(r'$p(x)$')
ax.yaxis.set_label_coords(-0.1,0.90)
#ax.set_ylim(np.min(y),np.max(y))

#ax.set_yscale('log')

# Tickers
ax.tick_params(axis='both', which='major', labelsize=12)

# Ajustar los mrgenes automticamente
plt.tight_layout()

#Legend
ax.legend(loc='upper right',bbox_to_anchor=(0.52,1),ncol=1,title='')

# Saving figure
plt.savefig('StandardNormal.pdf')
```
